```python
import pandas as pd

# 불러오기
df = pd.read_parquet("specpower.parquet")
df.head()

# 서버 카테고리 (mid만 남김)
print(df["server_category"].value_counts())

df = df[df["server_category"] == "mid"]

#컬럼 확인 및 제거
print(df.columns.to_list())

df = df.drop(columns = ["vendor", "system", "processor", "num_cores", "num_chips", "server_category"])

print(df)

# 내보내기
df.to_parquet("processed/spec.parquet", index=False)
df.to_csv("processed/spec.csv", index=False)
```
